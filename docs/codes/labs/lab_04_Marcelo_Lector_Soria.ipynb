**b)** Explore el archivo **`libertad_prensa_codigo.csv`** e identifique el código ISO que aparece asociado a dos nombres de país distintos. Elimine el registro que corresponda a un valor incorrecto o inconsistente, conservando solo el que considere válido.

**c)** Una vez preparados los archivos, cree un nuevo DataFrame llamado `df` que combine `df_anio` con `df_codigos`, utilizando la columna `codigo_iso` como clave. Asegúrese de realizar una unión que conserve únicamente los registros que tengan coincidencia en ambas fuentes.

> **Sugerencia**:
>
> * Para unir los archivos por filas (años), utilice la función `pd.concat([...])`.
> * Para combinar información por columnas (variables), utilice `pd.merge(...)` especificando `on='codigo_iso'`.



````python
archivos_anio = [
    'https://raw.githubusercontent.com/fralfaro/MAT281/main/docs/labs/data/libertad_prensa_01.csv',
    'https://raw.githubusercontent.com/fralfaro/MAT281/main/docs/labs/data/libertad_prensa_02.csv'
]

lista_dataframes = []
for archivo in archivos_anio:
    df_temporal = pd.read_csv(archivo)
    df_temporal.columns = df_temporal.columns.str.lower()
    lista_dataframes.append(df_temporal)

df_anio = pd.concat(lista_dataframes, ignore_index=True)

print("DataFrame 'df_anio' consolidado:")
print(df_anio.head())
````

````python
duplicados = df_codigos[df_codigos.duplicated(subset=['codigo_iso'], keep=False)]
print("\nCódigo ISO con nombres de país duplicados:")
print(duplicados)

df_codigos = df_codigos[df_codigos['pais'] != 'Republic of Zimbabwe']

print("\nVerificación post-limpieza para el código 'ZWE':")
print(df_codigos[df_codigos['codigo_iso'] == 'ZWE'])
````

````python
df = pd.merge(
    df_anio,
    df_codigos,
    on="codigo_iso",
    how="inner"
)

print("\nDataFrame final 'df' combinado:")
print(df.head())

print("\nResumen del DataFrame final:")
print(df.info())
````

En esta sección se busca identificar cuáles son los países de América Latina que han presentado los valores extremos del **Índice de Libertad de Prensa** en cada año observado.

#### **Tareas:**

* El país con el menor valor de `indice` (mayor libertad de prensa).
* El país con el mayor valor de `indice` (menor libertad de prensa).

**b)** Resuelve la misma tarea del punto anterior utilizando un enfoque vectorizado con `groupby`, sin usar ciclos explícitos.



#### **Lista de países latinoamericanos considerada:**

```python
america = ['ARG', 'ATG', 'BLZ', 'BOL', 'BRA', 'CAN', 'CHL', 'COL', 'CRI',
           'CUB', 'DOM', 'ECU', 'GRD', 'GTM', 'GUY', 'HND', 'HTI', 'JAM',
```

````python
# respuesta
america = ['ARG', 'ATG', 'BLZ', 'BOL', 'BRA', 'CAN', 'CHL', 'COL', 'CRI',
       'CUB', 'DOM', 'ECU', 'GRD', 'GTM', 'GUY', 'HND', 'HTI', 'JAM',
       'MEX', 'NIC', 'PAN', 'PER', 'PRY', 'SLV', 'SUR', 'TTO', 'URY',
       'USA', 'VEN']

df_america = df[df['codigo_iso'].isin(america)].copy()

# a) Con ciclo for:
print("Resultados con ciclo for:\n")
for anio in sorted(df_america['anio'].unique()):
    df_anio = df_america[df_america['anio'] == anio]

    if df_anio['indice'].dropna().empty:
        print(f"Año {anio}: no hay datos disponibles")
        print("-"*50)
        continue

    pais_min = df_anio.loc[df_anio['indice'].idxmin(), ['pais','indice']]
    pais_max = df_anio.loc[df_anio['indice'].idxmax(), ['pais','indice']]

    print(f"Año {anio}:")
    print(f"  Mayor libertad (menor índice): {pais_min['pais']} ({pais_min['indice']})")
    print(f"  Menor libertad (mayor índice): {pais_max['pais']} ({pais_max['indice']})")
    print("-"*50)


# b) Con groupby
print("\nResultados con groupby (vectorizado):\n")

df_america_clean = df_america.dropna(subset=['indice'])

idx_min = df_america_clean.groupby('anio')['indice'].idxmin()
minimos = df_america_clean.loc[idx_min, ['anio','pais','indice']].rename(columns={'pais':'pais_min','indice':'indice_min'})

idx_max = df_america_clean.groupby('anio')['indice'].idxmax()
maximos = df_america_clean.loc[idx_max, ['anio','pais','indice']].rename(columns={'pais':'pais_max','indice':'indice_max'})

extremos = pd.merge(minimos, maximos, on='anio')
print(extremos.head(15))
````

### 4. Análisis anual del índice por país

En esta sección se busca analizar la evolución del **índice máximo** de libertad de prensa alcanzado por cada país a lo largo del tiempo.

#### **Tarea principal:**

* Construye una tabla dinámica (`pivot_table`) donde las **filas** correspondan a los países, las **columnas** a los años (`anio`) y los **valores** sean el `indice` máximo alcanzado por cada país en ese año.
* Asegúrate de reemplazar los valores nulos resultantes con `0`.

> **Hint**: Puedes utilizar el parámetro `fill_value=0` en `pd.pivot_table(...)`.



#### **Preguntas adicionales:**

**a)** ¿Qué país tiene el mayor valor de `indice` en toda la tabla resultante? ¿Y cuál tiene el menor (distinto de cero)?
**b)** ¿Qué años presentan en promedio los valores de `indice` más altos? ¿Y los más bajos?

> (Pista: usa `.mean(axis=0)` sobre la tabla pivot)

**c)** ¿Qué país muestra mayor **variabilidad** (diferencia entre su máximo y mínimo `indice` a lo largo del tiempo)?

> (Pista: aplica `.max(axis=1) - .min(axis=1)`)

**d)** ¿Existen países con índice constante a lo largo de todos los años registrados? ¿Cuáles?

**e)** ¿Qué países no tienen ningún dato (es decir, quedaron con todos los valores igual a 0)? ¿Podrías explicar por qué?





````python
pivot = pd.pivot_table(
    df,
    values='indice',
    index='pais',
    columns='anio',
    aggfunc='max',
    fill_value=0
)

print("Tabla dinámica (primeros países):")
print(pivot.head())

# a)
max_val = pivot.values.max()
min_val = pivot.values[pivot.values > 0].min()

pais_max = pivot.max(axis=1).idxmax()
pais_min = pivot.replace(0, float('inf')).min(axis=1).idxmin()

print("\n[a]")
print(f"País con mayor índice en toda la tabla: {pais_max} ({max_val})")
print(f"País con menor índice (distinto de cero): {pais_min} ({min_val})")

# b)
promedios_anuales = pivot.mean(axis=0)

anio_max = promedios_anuales.idxmax()
anio_min = promedios_anuales.idxmin()

print("\n[b]")
print(f"Año con promedio más alto: {anio_max} ({promedios_anuales.max():.2f})")
print(f"Año con promedio más bajo: {anio_min} ({promedios_anuales.min():.2f})")

# c)
variabilidad = pivot.max(axis=1) - pivot.min(axis=1)
pais_var = variabilidad.idxmax()

print("\n[c]")
print(f"País con mayor variabilidad: {pais_var} ({variabilidad.max()})")

# d)
constantes = pivot[(pivot.max(axis=1) == pivot.min(axis=1))]

print("\n[d]")
if constantes.empty:
    print("No hay países con índice constante en todos los años.")
else:
    print("Países con índice constante en todos los años:")
    print(list(constantes.index))

# e)
sin_datos = pivot[(pivot == 0).all(axis=1)]

print("\n[e]")
if sin_datos.empty:
    print("Todos los países tienen al menos un dato.")
else:
    print("Países sin ningún dato (todos ceros):")
    print(list(sin_datos.index))
    print("Explicación: estos países no tuvieron registros en los archivos originales, "
          "por lo que la pivot_table los rellena con 0.")
````
